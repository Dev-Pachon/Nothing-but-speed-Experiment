```python
print('¿El conjunto de datos tiene nan?',df.isnull().values.any(),'\n')
print('¿Cuántos nan tiene en total?', df.isnull().sum().sum(),'\n')
#En cuáles de los atributos tenemos valores nan?
diccionario_de_nan = {}
for atributo in df.columns.values:
  diccionario_de_nan[atributo]=df[atributo].isnull().sum()
print('¿Cómo están distribuidos los nan?\n',diccionario_de_nan)
```

```python
#Debido a que en nuestro estudio las columnas usuario, fechaActualizacion, id y codigoProducto no tienen relevancia, las eliminaremos del dataframe
df = df.drop(columns="usuario")
df = df.drop(columns="fechaActualizacion")
df = df.drop(columns="id")
df = df.drop(columns="codigoProducto")
```

```python
def linePlot(dfPlot):
    fig1, ax = plt.subplots(1,3,figsize=(20,5))
    j=0  #Indexa la columna.
    for idx in range(6,9):
      atributo = str(dfPlot.columns[idx])
      title = 'Line plot\n'+atributo
      ax[j].set_title(title)
      sns.lineplot(ax=ax[j],x="fecha",y=atributo,data=dfPlot,linewidth=2)
      ax[j].set_xticklabels('')
      j+=1
```

```python
def boxPlot(dfPlot):
    fig1, ax = plt.subplots(1,3,figsize=(20,5))  #Creamos un subplot de 2 filas x 4 columnas.
    j=0  #Indexa la columna.
    for idx in range(6,9):
      atributo = str(dfPlot.columns[idx])
      title = 'Box plot\n'+atributo
      ax[j].set_title(title)
      ax[j].boxplot(dfPlot[atributo].values,
                      whis = 1.5)  
      ax[j].set_xticklabels('')
      j+=1
```

```python
rslt_df = df[df['grupo'] == "CARNICOS"]
linePlot(rslt_df)
boxPlot(rslt_df)
```
